```python
plt.figure(figsize=(16,5))
sns.scatterplot(data=df, x= "Month", y= "Total", color="c")
plt.title("Month wise Sales")
#plt.savefig("Month wise Sales")
plt.show()
```

In Scatterplot you can see more sales are achived in Jan, Feb and Mar.

```python
plt.figure(figsize=(16,5))
sns.scatterplot(data=df, x= "Month", y= "Total", hue= "Gender", color="c")
plt.title("Month wise Sales by Gender")
#plt.savefig("Month wise Sales by Gender")
plt.show()
```

In Scatterplot you can see more sales are achived in Jan, Feb and Mar by Gender

```python
sns.countplot(data=df, x="City", hue="Gender")
plt.title("Sales Count by Gender in City")
#plt.savefig("Sales Count by Gender in City")
plt.show()
```

In the count plot it can be seen that majority of the customer in Delhi is Female Population and in Banglore majority is from the Male Population

## Plotting for different Payments methods count

```python
payment_df = pd.DataFrame(df["Payment"].value_counts())
payment_df
```

```python
sns.barplot(x= payment_df.index, y= payment_df.Payment)
plt.title("Overall Region Mode of Payment")
#plt.savefig("Overall Region Mode of Payment")
plt.show()
```

The Bar show in overall region that the Payment recieved is more from Ewallet then Cash then Credit Card.

```python
sns.countplot(data=df, x="Gender", hue="Payment",palette="hot")
plt.title("Overall Region Mode of Payment by Gender")
#plt.savefig("Overall Region Mode of Payment by Gender")
plt.show()
```

Most Males prefere to pay by cash and most Females prefere to pay by Ewallet in overall Region.

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Region", y= "Total", palette="hot")
plt.title("Total Sales by Region")
#plt.savefig("Total Sales by Region")
plt.show()
```

This Bar Plot shows that the Majority of Sales achived from South Region.

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Region", y= "Total", hue="Gender", palette="hot")
plt.title("Total Sales by Region by Gender")
#plt.savefig("Total Sales by Region by Gender")
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Month", y= "Total", hue="Gender", palette="hot")
plt.title("Total Monthly Sales by Gender")
#plt.savefig("Total Monthly Sales by Gender")
plt.show()
```

This barplot shows Month on Month Sales for Male and Female population from all Region.

 ## Creating Dummies for Gender Column to get numerical data for Seaborn Barplot.

```python
gender_dummies= pd.get_dummies(df["Gender"])
gender_dummies.head()
```

## Joining Original Data Frame (df) with gender dummies data frame (gender_dummies).

```python
df = pd.concat([df, gender_dummies], axis=1)
df.head()
```

## For Male

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Male", y= "Product line")
plt.title("Male Total Sales by Product Line")
#plt.savefig("Male Total Sales by Product Line")
plt.show()
```

It can be seen that the Majority of Male Population Buys Fassion Accessories & Sports and travel from Product Line.

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Customer type", y= "Male", hue="City")
plt.title("Male Customer Type by City")
#plt.savefig("Male Customer Type by City")
plt.show()
```

Major Males Customer from Banglore are of Customer type Member & Normal both.

```python
plt.figure(figsize=(16,5))
sns.lineplot(data=df, x= "Male", y= "Rating", hue="City")
plt.title("Male Rating by City")
#plt.savefig("Male Rating by City")
plt.show()
```

The Ratings in Banglore and Mumbai are seen on the higher side and lower in delhi from Male Population.

## For Female

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Female", y= "Product line")
plt.title("Female Total Sales by Product Line")
#plt.savefig("Female Total Sales by Product Line")
plt.show()
```

It can be seen that the Majority of Female Population Buys Health and Beauty Product from Product Line.

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x= "Customer type", y= "Female", hue="City")
plt.title("Female Customer Type by City")
#plt.savefig("Female Customer Type by City")
plt.show()
```

Major Females Customer from Delhi are of Customer type Member & Normal both.

```python
plt.figure(figsize=(16,5))
sns.lineplot(data=df, x= "Female", y= "Rating", hue="City")
plt.title("Female Rating by City")
#plt.savefig("Female Rating by City")
plt.show()
```

The Ratings in Banglore and Mumbai are seen on the lower side and higher in delhi from Female Population.
